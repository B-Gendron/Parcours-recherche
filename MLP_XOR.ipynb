## Implémentation d'un perceptron multi-couches (2 couches ici) pour apprendre XOR.

Pour commencer, on rappelle que XOR, le "ou exclusif", est défini par la table de vérité suivante :
$$
\begin{array}{|c|c|c|}
\hline
\text{XOR} & x_1=0 & x_1 = 1 \\
\hline
x_2 = 0 & 0 & 1 \\
\hline
x_2 = 1 & 1 & 0 \\
\hline
\end{array}$$

Dans cette étude, on va utiliser un réseau de neurones pour prédire la valeur de $\text{XOR}(x_1, x_2)$ avec $x_1$ et $x_2$ donnés en input.
